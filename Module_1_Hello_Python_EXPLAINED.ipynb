## 1️⃣ PRINT() - Tu Primera Línea de Código

## 2️⃣ VARIABLES - "Cajas" para Guardar Datos

### ¿Qué es una variable?
Una **variable** es un contenedor que guarda un valor.

Piénsalo como una **caja etiquetada**:
- El **nombre** de la caja es el nombre de la variable
- Lo **dentro** es el valor
- Usamos `=` para poner algo dentro

### Sintaxis:
```python
nombre_variable = valor
```

### Ejemplo en Tu Industria:
En lugar de escribir "480" cada vez (minutos de un turno), guardas eso en una variable.

````python
# Crear variables para un análisis de producción
turno_minutos = 480          # Un turno tiene 480 minutos (8 horas)
tiempo_parada = 120          # La máquina estuvo parada 120 minutos
produccion_piezas = 1500     # Se produjeron 1500 piezas

print(turno_minutos)
print(tiempo_parada)
print(produccion_piezas)
````

### ¿Por qué usar variables?
- **Claridad:** El código es legible
- **Reutilización:** Usas la variable varias veces
- **Mantenimiento:** Si cambian datos, cambias 1 lugar, no 10

## 3️⃣ TIPOS DE DATOS - Los 4 Principales

Python tiene diferentes **tipos** de datos. Los 4 principales:

| Tipo | Ejemplo | Explicación |
|------|---------|-------------|
| `int` | 42 | Números **sin** decimales |
| `float` | 3.14 | Números **con** decimales |
| `str` | "Hola" | **Texto** (siempre entre comillas) |
| `bool` | True/False | **Verdadero o Falso** |

### ¿Por qué importa?
Porque Python trata cada tipo diferente. No puedes sumar "texto" + número sin problemas.

````python
# TIPO 1: int (enteros)
defectos = 5              # 5 piezas defectuosas
print(f"Defectos: {defectos}")
print(f"Tipo: {type(defectos)}")
````

````python
# TIPO 2: float (decimales)
oee_porcentaje = 85.5     # OEE es 85.5%
print(f"OEE: {oee_porcentaje}%")
print(f"Tipo: {type(oee_porcentaje)}")
````

````python
# TIPO 3: str (texto)
linea_produccion = "Línea A"  # Nombre de la línea
print(f"Línea: {linea_produccion}")
print(f"Tipo: {type(linea_produccion)}")
````

````python
# TIPO 4: bool (verdadero/falso)
maquina_activa = True      # ¿Está activa la máquina?
print(f"¿Máquina activa?: {maquina_activa}")
print(f"Tipo: {type(maquina_activa)}")
````

## 4️⃣ OPERACIONES ARITMÉTICAS - Matemáticas en Python

Python hace cálculos como una calculadora.

| Operador | Significado | Ejemplo |
|----------|-------------|----------|
| `+` | Suma | 10 + 5 = 15 |
| `-` | Resta | 10 - 5 = 5 |
| `*` | Multiplicación | 10 * 5 = 50 |
| `/` | División | 10 / 5 = 2.0 |
| `**` | Potencia | 2 ** 3 = 8 |
| `//` | División entera | 10 // 3 = 3 |
| `%` | Módulo (resto) | 10 % 3 = 1 |

````python
# EJEMPLO REAL: Calcular Disponibilidad de Equipamiento
# Disponibilidad = (Tiempo Total - Tiempo de Parada) / Tiempo Total

tiempo_total = 480        # minutos en un turno
tiempo_parada = 60        # minutos que estuvo parada

tiempo_productivo = tiempo_total - tiempo_parada
disponibilidad = tiempo_productivo / tiempo_total
disponibilidad_porcentaje = disponibilidad * 100

print(f"Tiempo total: {tiempo_total} minutos")
print(f"Tiempo parada: {tiempo_parada} minutos")
print(f"Tiempo productivo: {tiempo_productivo} minutos")
print(f"Disponibilidad: {disponibilidad_porcentaje:.1f}%")
````

## 5️⃣ COMPARACIONES - Preguntas que Python Responde

### ¿Cuándo usas comparaciones?
Cuando necesitas hacer preguntas del tipo:
- ¿Es A mayor que B?
- ¿Son A y B iguales?
- ¿Es A menor o igual que B?

### Operadores de Comparación:

| Operador | Significado | Ejemplo |
|----------|-------------|----------|
| `==` | ¿Igual a? | 5 == 5 → True |
| `!=` | ¿Diferente a? | 5 != 3 → True |
| `>` | ¿Mayor que? | 5 > 3 → True |
| `<` | ¿Menor que? | 5 < 3 → False |
| `>=` | ¿Mayor o igual? | 5 >= 5 → True |
| `<=` | ¿Menor o igual? | 5 <= 5 → True |

### ⚠️ IMPORTANTE: `=` vs `==`
- `=` **ASIGNA** un valor (guarda en variable)
- `==` **COMPARA** valores (pregunta si son iguales)

````python
# EJEMPLO REAL: ¿El OEE cumple el objetivo?
oee_actual = 85.5
oee_objetivo = 85.0

print(f"OEE Actual: {oee_actual}")
print(f"OEE Objetivo: {oee_objetivo}")
print()
print(f"¿OEE cumple objetivo? {oee_actual >= oee_objetivo}")
print(f"¿OEE es mejor que objetivo? {oee_actual > oee_objetivo}")
````

## 6️⃣ CONDICIONALES (if/else) - Tomar Decisiones

### ¿Qué es un condicional?
Es una **decisión** que Python toma basada en una pregunta.

**Lógica:**
- IF (Si) la condición es TRUE → ejecuta esto
- ELSE (Si no) → ejecuta esto otro

### Sintaxis:
```python
if condición:
    # código si es verdadero
else:
    # código si es falso
```

### ⚠️ IMPORTANTE: Indentación
En Python, los espacios **importan**. El código dentro del `if` debe estar **indentado** (con espacios al inicio).

````python
# EJEMPLO REAL: ¿Alarma por baja calidad?
defectos = 10
limite_critico = 15  # Si hay más de 15 defectos, ALERTA

if defectos > limite_critico:
    print("⚠️ ALERTA: Demasiados defectos! Revisar proceso.")
else:
    print("✅ Calidad dentro de límites normales")
````

````python
# EJEMPLO 2: Clasificación de OEE
oee = 78.5

if oee >= 85:
    print("✅ Excelente")
else:
    print("⚠️ Por mejorar")
````

## 7️⃣ FUNCIONES - Código Reutilizable

### ¿Qué es una función?
Una **función** es un bloque de código que:
1. Tiene un nombre
2. Puede recibir **parámetros** (datos que le pasas)
3. Hace algo
4. Devuelve un **resultado**

### ¿Por qué usarlas?
En lugar de escribir el mismo código 100 veces, lo escribes 1 vez en una función y lo reutilizas.

### Sintaxis:
```python
def nombre_funcion(parámetros):
    # código
    return resultado
```

````python
# FUNCIÓN REAL: Calcular Disponibilidad
def calcular_disponibilidad(tiempo_total, tiempo_parada):
    """Calcula la disponibilidad del equipamiento"""
    tiempo_productivo = tiempo_total - tiempo_parada
    disponibilidad = (tiempo_productivo / tiempo_total) * 100
    return disponibilidad

# Usar la función
disp_linea_a = calcular_disponibilidad(480, 60)  # 8 horas, 1 hora parada
disp_linea_b = calcular_disponibilidad(480, 120) # 8 horas, 2 horas paradas

print(f"Disponibilidad Línea A: {disp_linea_a:.1f}%")
print(f"Disponibilidad Línea B: {disp_linea_b:.1f}%")
````

## 8️⃣ LISTAS - Colecciones de Datos

### ¿Qué es una lista?
Una **lista** es un contenedor que guarda **múltiples valores**.

**Sintaxis:**
```python
mi_lista = [valor1, valor2, valor3]
```

### ¿Por qué usarlas?
En lugar de crear 100 variables para 100 valores, los guardas en 1 lista.

### Índices (Posiciones):
Python empieza a **contar desde 0**:
```
[10, 20, 30, 40]
 0   1   2   3
```

````python
# EJEMPLO REAL: Defectos por día de la semana
defectos_semana = [12, 15, 8, 10, 18, 14, 9]
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]

print(f"Defectos Lunes: {defectos_semana[0]}")
print(f"Defectos Martes: {defectos_semana[1]}")
print(f"Máximo defectos: {max(defectos_semana)}")
print(f"Mínimo defectos: {min(defectos_semana)}")
print(f"Promedio defectos: {sum(defectos_semana) / len(defectos_semana):.1f}")
````

## 9️⃣ LOOPS (for) - Repetir Código

### ¿Qué es un loop?
Un **loop** es un código que se **repite**.

En lugar de escribir `print()` 7 veces, usas un loop que hace `print()` 7 veces automáticamente.

### Sintaxis:
```python
for variable in lista:
    # código que se repite
```

### ¿Cuándo usarla?
- Procesar cada elemento de una lista
- Repetir una acción N veces

````python
# EJEMPLO REAL: Mostrar defectos de cada día
defectos_semana = [12, 15, 8, 10, 18, 14, 9]
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]

for i in range(len(dias)):
    print(f"{dias[i]}: {defectos_semana[i]} defectos")
````

````python
# FORMA MÁS SIMPLE: Loop directo sobre la lista
defectos_semana = [12, 15, 8, 10, 18, 14, 9]

for defectos in defectos_semana:
    if defectos > 15:
        print(f"⚠️ Muchos defectos: {defectos}")
    else:
        print(f"✅ Defectos aceptables: {defectos}")
````
