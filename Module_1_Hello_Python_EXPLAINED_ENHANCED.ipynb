## ➕ EXTRA 1: COMENTARIOS vs DOCSTRINGS - ¿Cuál Usar?

### Tipos de comentarios en Python

Hay dos formas de explicar código:

**1. Comentarios NORMALES (`#`)** - Solo para humanos
```python
# Este comentario explica el código
# Python lo IGNORA completamente
x = 10  # Comentario al lado del código
```

**2. Docstrings (`"""..."""`)** - Documentación ESPECIAL que Python GUARDA
```python
def mi_funcion():
    """Esta es la documentación de la función.
    Python la GUARDA y puedes acceder con help().
    """
    return 42
```

### Diferencia IMPORTANTE

| Aspecto | Comentario `#` | Docstring `""""""` |
|---------|----------------|--------------------|
| Propósito | Explicar código complicado | Documentar funciones/clases |
| Python lo guarda | NO | SÍ |
| Se puede acceder con `help()` | NO | SÍ |
| Ubicación típica | Cualquier lado | Primero dentro de función |
| Ejemplo | `# Calculamos OEE` | `"""Calcula el OEE"""` |

````python
# EJEMPLO REAL: Función con comentarios Y docstring

def calcular_disponibilidad(tiempo_total, tiempo_parada):
    """Calcula la disponibilidad del equipamiento.
    
    Parámetros:
        tiempo_total (int): Tiempo total del turno en minutos
        tiempo_parada (int): Tiempo que estuvo parada la máquina
    
    Retorna:
        float: Porcentaje de disponibilidad (0-100)
    
    Ejemplo:
        >>> calcular_disponibilidad(480, 60)
        87.5
    """
    # Calculamos el tiempo productivo (comentario normal)
    tiempo_productivo = tiempo_total - tiempo_parada
    
    # Calculamos disponibilidad como porcentaje
    disponibilidad = (tiempo_productivo / tiempo_total) * 100
    
    return disponibilidad

# Ahora puedes ver la documentación con help():
help(calcular_disponibilidad)
````

## ➕ EXTRA 2: F-STRINGS - Mezclar Texto con Variables

### ¿Qué es un f-string?

`f"..."` = "formatted string" (cadena formateada)

Permite meter **variables dentro de texto** usando `{}`.

### Evolución: 3 formas de hacer lo mismo

````python
disp_a = 87.5

# FORMA 1: Concatenación (VIEJA, complicada)
print("Disponibilidad Línea A: " + str(disp_a) + "%")

# FORMA 2: .format() (MENOS vieja)
print("Disponibilidad Línea A: {}%".format(disp_a))

# FORMA 3: f-string (NUEVA, la mejor)
print(f"Disponibilidad Línea A: {disp_a}%")
````

### ¿Por qué f-strings?

✅ **Más legible** - El código se entiende de una mirada  
✅ **Más rápido** - Python lo procesa más eficientemente  
✅ **Menos errores** - No necesitas `str()` para convertir tipos  

### Ejemplo lado a lado

````python
linea = "Línea A"
defectos = 10
oee = 85.5

# SIN f-string (difícil de leer)
print(linea + ": " + str(defectos) + " defectos, OEE: " + str(oee) + "%")

# CON f-string (mucho más limpio)
print(f"{linea}: {defectos} defectos, OEE: {oee}%")
````

## ➕ EXTRA 3: Formatos Dentro de F-Strings (`:` )

### ¿Qué es `:.1f`?

Cuando ves código como `{variable:.1f}`, la parte después de `:` es un **formato**.

El formato **CONTROLA CÓMO SE MUESTRA** el número.

### Ejemplo: Decimales

````python
valor = 87.5234567  # Muchos decimales

print("Sin formato:")
print(f"OEE: {valor}%")
print()

print("Con formatos diferentes:")
print(f"Sin decimales:     {valor:.0f}")     # 88
print(f"1 decimal:         {valor:.1f}")     # 87.5
print(f"2 decimales:       {valor:.2f}")     # 87.52
print(f"3 decimales:       {valor:.3f}")     # 87.523
````

### Otros formatos útiles

````python
numero = 1234.5
dinero = 50000

print("Diferentes formatos:")
print(f"Normal:              {numero}")
print(f"Con separador mil:   {numero:,.1f}")
print(f"Porcentaje:          {0.855:.1%}")     # 85.5%
print(f"Con $ (dinero):      ${dinero:,.2f}")  # $50,000.00
````

### Guía de Formatos Comunes

| Formato | Ejemplo | Resultado |
|---------|---------|----------|
| `:.0f` | `{87.5:.0f}` | `88` (sin decimales) |
| `:.1f` | `{87.5:.1f}` | `87.5` (1 decimal) |
| `:.2f` | `{87.5:.2f}` | `87.50` (2 decimales) |
| `:.1%` | `{0.855:.1%}` | `85.5%` (porcentaje) |
| `:,` | `{1000:,}` | `1,000` (separador mil) |
| `:,.2f` | `{1000.5:,.2f}` | `1,000.50` (mil + decimales) |

## ➕ EXTRA 4: EJEMPLO PRÁCTICO COMPLETO

Combina todo: comentarios, docstring, f-strings, formatos

````python
def calcular_oee(disponibilidad, rendimiento, calidad):
    """Calcula el OEE (Overall Equipment Effectiveness).
    
    Formula: OEE = Disponibilidad × Rendimiento × Calidad × 100
    
    Parámetros:
        disponibilidad (float): % de disponibilidad (0-1)
        rendimiento (float): % de rendimiento (0-1)
        calidad (float): % de calidad (0-1)
    
    Retorna:
        float: OEE en porcentaje (0-100)
    """
    # Calculamos OEE
    oee = disponibilidad * rendimiento * calidad * 100
    return oee

# Datos de ejemplo
disponibilidad = 0.875  # 87.5%
rendimiento = 0.95      # 95%
calidad = 0.98          # 98%

# Calcular OEE
oee_resultado = calcular_oee(disponibilidad, rendimiento, calidad)

# Mostrar resultado con formato profesional
print(f"\n{'='*50}")
print(f"ANÁLISIS DE EFICIENCIA - LÍNEA A")
print(f"{'='*50}")
print(f"Disponibilidad: {disponibilidad:.1%}")
print(f"Rendimiento:    {rendimiento:.1%}")
print(f"Calidad:        {calidad:.1%}")
print(f"{'='*50}")
print(f"OEE TOTAL:      {oee_resultado:.2f}%")
print(f"{'='*50}")
````

| Operador | Significado | Ejemplo |
|----------|------------|----------|

## 9️⃣ LOOPS (for) - Repetir Código

```python
for variable in lista:
    # código que se repite
```

````python
# EJEMPLO: Mostrar defectos con alerta
defectos_semana = [12, 15, 8, 10, 18, 14, 9]
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]

for i in range(len(dias)):
    if defectos_semana[i] > 15:
        print(f"{dias[i]}: {defectos_semana[i]} ALERTA!")
    else:
        print(f"{dias[i]}: {defectos_semana[i]} OK")
````
